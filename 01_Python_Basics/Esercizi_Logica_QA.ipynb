## Esercizio 1: Validazione email¶
**Scenario:** Verificare se un'email inserita è valida oppure no (scartare le email che non contengono la @ o che non finiscono con .com o .it)

```python
# Aggiungo nell'email degli spazi per testare la pulizia (strip) e converto in minuscolo, per evitare che un test fallisca per un errore di battituta
# strip() -> è fondamentale perchè spesso i bug nei test di automazione nascono perchè "catturiamo" dal sito uno spazio vuoto invisibile

email_inserita = "  test.utente@gmail.com "
email = email_inserita.strip().lower()

if "@" in email and email.endswith((".com", ".it")):   # .endswith() -> metodo che permette di controllare più estensioni contemporaneamente passandole all'interno di una parentesi tonda (una "tupla")
    print("TEST PASSATO: l'indirizzo email", email, "è in un formato valido")
elif "@" not in email:
    print("TEST FALLITO: Manca il simbolo '@' nell'email.")
else:
    print("TEST FALLITO: L'estensione finale non è supportata (usa .com o .it)")
        
```
