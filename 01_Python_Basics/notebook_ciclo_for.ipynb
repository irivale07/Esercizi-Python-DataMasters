# Ciclo FOR

```python
for i in range(0,10):
    print("la variabile i vale: ", i)

print("\n")
for i in range(0, 10, 2):
    print("la variabile i vale: ", i)
        
```

```python
min = -10
max = 10

print ("I numeri nell'intervallo -10, 10 sono: ")
for i in range(min, max + 1):
    print(i)
    
```

```python
cont = 0 #aggiungo la variagile contatore

for i in range (min, max + 1):
    if i % 2 == 0:   #controllo che il numero sia pari
        print(i, "pari")
        cont = cont + 1

    if cont >= 5:
        print("Ho trovato i primi 5 numeri pari nell'intervallo, esco")
        break
```

```python
s = "Lorem ipsum dolor sit amet"
contatore_m = 0

for letter in s:
    if letter == "m":
        contatore_m = contatore_m + 1

print("La lettera 'm' appare", contatore_m, "volte")
```
